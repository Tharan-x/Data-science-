```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,79,29)
y=np.sin(x)
plt.plot(x, y, label="Sine Wave", color='blue')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

data = np.random.randn(1000)
df = pd.DataFrame(data, columns=['Values'])

sns.histplot(df['Values'], bins=30, kde=True, color='blue')
plt.title("Histogram with KDE")
plt.show()
tips = sns.load_dataset('tips')
plt.figure(figsize=(6, 4))
sns.boxplot(x=tips['total_bill'])
plt.title("Box Plot of Total Bill")
plt.show()
sns.pairplot(tips, hue='sex')
plt.show()

corr_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

df = pd.DataFrame({
    "x": np.linspace(0, 10, 100),
    "y": np.sin(np.linspace(0, 10, 100))
})

fig = px.line(df, x='x', y='y', title="Interactive Sine Wave")
fig.show()
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', size='size', title="Total Bill vs Tip")
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import seaborn as sns

# Load Seaborn dataset for Scatter and 3D plots
tips = sns.load_dataset('tips')

# 3.1 Interactive Line Plot
df = pd.DataFrame({
    "x": np.linspace(0, 10, 100),
    "y": np.sin(np.linspace(0, 10, 100))
})

fig = px.line(df, x='x', y='y', title="Interactive Sine Wave")
fig.show()

# 3.2 Interactive Scatter Plot
fig = px.scatter(
    tips, x='total_bill', y='tip', color='sex', size='size',
    title="Total Bill vs Tip"
)
fig.show()

# 3.3 Interactive 3D Scatter Plot
fig = go.Figure(data=[
    go.Scatter3d(
        x=tips['total_bill'],
        y=tips['tip'],
        z=tips['size'],
        mode='markers',
        marker=dict(size=5, color=tips['total_bill'], colorscale='Viridis')
    )
])

fig.update_layout(title="3D Scatter Plot of Total Bill, Tip & Size")
fig.show()
```
